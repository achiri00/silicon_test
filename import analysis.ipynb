```python
leading_digits = exp['Custom Value (USD)'].astype(str).str[0].astype(int)
```

```python
benfords_law = np.log10(1 + 1 / np.arange(1, 10))
benfords_law *= len(leading_digits)
```

```python
# Plot actual vs. expected distribution
plt.bar(range(1, 10), np.histogram(leading_digits, bins=np.arange(1, 11))[0], alpha=0.5, label='Actual')
plt.plot(range(1, 10), benfords_law, 'ro-', label="Benford's Law")
plt.title("Benford's Law Analysis")
plt.xlabel('Leading Digit')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
benfords_law = np.log10(1 + 1 / np.arange(1, 10))
benfords_law_percentages = benfords_law / benfords_law.sum() * 100
for digit, percentage in zip(range(1, 10), benfords_law_percentages):
    print(f"Digit {digit}: {percentage:.2f}%")
```

```python
exp_2014 = exp[exp['Year'] == 2014]
exp_2015 = exp[exp['Year'] == 2015]

export_2014 = exp_2014.groupby('Industry')['Custom Value (USD)'].sum()
export_2015 = exp_2015.groupby('Industry')['Custom Value (USD)'].sum()

# Plot the data
fig, ax = plt.subplots(figsize=(10, 6))

# Bar plot for 2014 exports in blue
ax.bar(export_2014.index, export_2014, color='blue', label='2014')

# Bar plot for 2015 exports in orange
ax.bar(export_2015.index, export_2015, color='orange', label='2015')

ax.set_xlabel('Industry')
ax.set_ylabel('Custom Value (USD)')
ax.set_title('Exports in 2014 and 2015 by Industry')
ax.legend()

plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
